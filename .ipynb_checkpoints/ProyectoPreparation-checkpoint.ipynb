```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from scipy import stats
```

Importamos la base de datos y lo ponemos en un data frame

```python
df = pd.read_csv('Publications17-21 (1).xlsx')
df
```

Hacemos una deteccion de valores ausentes y los reemplazamos por el promedio

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 13', inplace=True, axis=1) 
df.drop('Unnamed: 14', inplace=True, axis=1) 
for column in df: #Iteramos cada columna, sacamos su promedio y reemplazamos los valores faltantes con el promedio

    avg = df[column].mean()
    df[column].fillna(avg, inplace=True)
```

```python
#Verificamos que ya no existan valores sin datos
df.isnull().sum()
```

Detectamos valores atipicos, los eliminamos

```python
df.describe()
```

```python
df = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)] #Usamos una libreria donde si existe un dato con un zscore mayor a 3 (Muy fuera) Se elimina
```

```python
df.describe()
```
